```python
from collections import defaultdict
sample = [1,3,2,5,7,0,2,3]
def mode(values):
    counts = defaultdict(lambda:0)
    for s in values:
        counts[s] += 1
        
    max_count = max(counts.values())
    modes = [v for v in set(values) if counts[v] == max_count]  #set(values) creates a set of unique elements in the list values and uses v to iterate over them to check if counts[v] frequency is equal to the maxcount
    return modes

print(mode(sample))
```

```python
# retrieving a critical z value 

from scipy.stats import norm
def critical_z_value(p):
    norm_dist = norm(loc = 0.0 , scale = 1.0)
    left_tail_area = (1.0-p)/2.0
    upper_area = 1.0 - ((1.0-p)/2.0)
    return norm_dist.ppf(left_tail_area),norm_dist.ppf(upper_area)

print(critical_z_value(p = .95))
```

```python
from math import sqrt
from scipy.stats import norm

def critical_z_value(p):
    norm_dist = norm(loc = 0.0 , scale = 1.0)
    left_tail_area = (1.0-p)/2.0
    upper_area = 1.0 - ((1.0-p)/2.0)
    return norm_dist.ppf(left_tail_area),norm_dist.ppf(upper_area)

def confidence_interval(p, sample_mean , sample_std , n):
    lower , upper = critical_z_value(p)
    print(lower , upper)
    lower_ci = lower*(sample_std/sqrt(n))
    upper_ci = upper*(sample_std/sqrt(n))
    print(lower_ci , upper_ci)
    return sample_mean + lower_ci, sample_mean + upper_ci

print(confidence_interval(p = 0.95 , sample_mean = 64.408 , sample_std = 2.05 , n = 31))
    
```

```python
# exercise 3 from ch 3
from scipy.stats import norm
def critical_z_values(p, mean = 0.0 , std = 1.0):
    norm_dist = norm(loc = mean , scale = std)
    left_area = (1.0-p)/2.0
    right_area = 1-((1.0-p)/2.0)
    return norm_dist.ppf(right_area),norm_dist.ppf(left_area)

n = 34
sample_mean = 1.715588
sample_std_dev = 0.029252
upper , lower = critical_z_values(0.99)
upper_ci = upper * (sample_std_dev / sqrt(n))
lower_ci = lower * (sample_std_dev / sqrt(n))
print("{0} - {1}".format(sample_mean + lower_ci , sample_mean + upper_ci))
```

```python
import pandas as pd
points = list(pd.read_csv("https://bit.ly/2KF29Bd").itertuples())
m = 0.0
b = 0.0
L = .001
iterations = 100000
n = float(len(points))
for i in range(iterations):
    D_m = sum(2 * p.x * ((m * p.x + b) - p.y) for p in points)
    D_b = sum(2 * ((m* p.x + b ) - p.y) for p in points)
    
    m -= L * D_m
    b -= L * D_b
    
print(m , b)
```

```python
import pandas as pd
from sympy import *
points = list(pd.read_csv("https://bit.ly/2KF29Bd").itertuples())
m, b , i , n = symbols('m b i n')
x,y = symbols('x y', cls = Function)
sum_of_squares = Sum((m*x(i) + b - y(i))**2,(i , 0 , n))
d_m = diff(sum_of_squares,m)\
    .subs(n , len(points) - 1).doit()\
    .replace(x , lambda i : points[i].x)\
    .replace(y , lambda i : points[i].y)

d_b = diff(sum_of_squares,b)\
    .subs(n , len(points) - 1).doit()\
    .replace(x , lambda i : points[i].x)\
    .replace(y , lambda i : points[i].y)

d_m = lambdify([m,b],d_m)
d_b = lambdify([m,b],d_b)

m = 0.0
b = 0.0

L = 0.001

iterations = 100_000
for i in range(iterations):
    m -= L*d_m(m,b)
    b -= L*d_b(m,b)
    
print(m , b)
```

```python
import pandas as pd
import numpy as np
data = pd.read_csv("https://bit.ly/2KF29Bd" , header = 0)

X = data.iloc[:,0].values
Y = data.iloc[:,1].values

n = data.shape[0]
m = 0.0
b = 0.0
sample_size = 1
L = 0.0001
epochs = 1_000_000
for i in range(epochs):
    idx = np.random.choice(n , sample_size, replace=False)
    x_sample = X[idx]
    y_sample = Y[idx]
    Y_pred = m*x_sample + b
    D_m = (-2/sample_size)*sum(x_sample*(y_sample - Y_pred))
    D_b = (-2/sample_size)*sum(y_sample - Y_pred)
    m -= L*D_m
    b -= L*D_b
    if i%100000 == 0:
        print(i , m , b)
        
print(m, b)
    
```

```python
from scipy.stats import t
from math import sqrt
n = 10
lower_cv = t(n-1).ppf(0.025)
upper_cv = t(n-1).ppf(0.975)

r = 0.957586

test_value = r/sqrt((1-r**2)/(n-2))

if test_value < lower_cv or test_value > upper_cv:
    print("Reject Null Hypothesis, as correlation proven")
else:
    print("Correlation not proven, couldn't reject null hypothesis")
    
if test_value > 0:
    p_value = 1.0 - t(n-1).cdf(test_value)
else:
    p_value = t(n-1).cdf(test_value)
    
p_value = p_value * 2
print("p_value = {0}".format(p_value))
```

```python
import pandas as pd
from scipy.stats import t
from math import sqrt

points = list(pd.read_csv('https://bit.ly/2KF29Bd', delimiter=",").itertuples())

m = 1.939
b = 4.733
n =len(points)
x_0 = 8.5
x_mean = sum(p.x for p in points)/n
t_value = t(n-2).ppf(.975)
standard_error = sqrt(sum((p.y - (m*p.x + b))**2 for p in points)/(n-2))
margin_of_error = t_value*standard_error * sqrt(1+(1/n)+(n*(x_0 - x_mean)**2)/(n * sum(p.x**2 for p in points) - sum(p.x for p in points)**2))

predicted_y = m*x_0 + b
print(predicted_y - margin_of_error, predicted_y + margin_of_error )
```

```python
import matplotlib.pyplot as plt
from scipy.stats import t
import numpy as np
fig, ax = plt.subplots()

degrees_of_freedom = [1, 5, 15, 30]  # Different df values to compare

x = np.linspace(-4, 4, 200)  # Range of values for the distributions

for df in degrees_of_freedom:
    y = t.pdf(x, df)  # Calculate the probability density
    ax.plot(x, y, label=f"df = {df}")

ax.set_xlabel("t-value")
ax.set_ylabel("Probability Density")
ax.set_title("t-distributions with Varying Degrees of Freedom")
ax.legend()
plt.show()
```

```python
import math
import pandas as pd

patient_data = pd.read_csv('https://bit.ly/33ebs2R', delimiter=',').itertuples()

b0 = -3.17576395
b1 = 0.69267212

def logistic_function(x):
    p = 1.0 / (1 + math.exp(-(b0 + b1 * x)))
    return p

joint_likelihood = 0.0

for p in patient_data:
    likelihood = (logistic_function(p.x) ** p.y) * ((1.0 - logistic_function(p.x)) ** (1.0 - p.y))
    joint_likelihood += math.log(likelihood)

joint_likelihood = math.exp(joint_likelihood)

print(joint_likelihood)
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

employee_data = pd.read_csv("https://tinyurl.com/y67qjrp")
inputs = employee_data.iloc[:, :-1]
output = employee_data.iloc[:, -1]

fit = LogisticRegression(penalty='none').fit(inputs, output)

print("COEFFICIENTS: {0}".format(fit.coef_.flatten()))
print("INTERCEPT: {0}".format(fit.intercept_.flatten()))

def predict_employee_will_stay(sex, age, promotions, years_employed):
    prediction = fit.predict([[sex, age, promotions, years_employed]])
    probabilities = fit.predict_proba([[sex, age, promotions, years_employed]])

    if prediction[0] == 1:
        return "WILL LEAVE, Probabilities: {0}".format(probabilities)
    else:
        return "WILL STAY, Probabilities: {0}".format(probabilities)

while True:
    n = input("Predict employee will stay or leave {sex}, {age}, [promotions}, {years employed}: ")
    (sex, age, promotions, years_employed) = n.split(",")
    print(predict_employee_will_stay(int(sex), int(age), int(promotions), int(years_employed)))
```
